```python
res = []
for user in tqdm(level_iron['summonerName']):
    url = f"https://kr.api.riotgames.com/lol/summoner/v4/summoners/by-name/{user}"
    puuid = requests.get(url, headers = request_header)
    if puuid.status_code == 429:
        print("Limit")
        print(user)
        break
    elif puuid.status_code == 404:
        time.sleep(1)
        continue
    else:
        puuid = puuid.json()
        url_2 = f"https://asia.api.riotgames.com/lol/match/v5/matches/by-puuid/{puuid['puuid']}/ids?type=ranked&start=0&count=10"
        matchid = requests.get(url_2, headers = request_header)
        if matchid.status_code == 429:
            print("Limit")
            break
        elif matchid.status_code == 404:
            time.sleep(1)
            continue
        else:
            res += matchid.json()
            time.sleep(1)
```

```python
match = pd.DataFrame(res)
match.to_csv(f"iron_match.csv", index = False, encoding = 'cp949')
```
