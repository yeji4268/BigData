## 아이언 티어 소환사 데이터 전처리
---
### 데이터 불러오기

```python
data_iron = []
for division in ['I', 'II', 'III', 'IV']:
    data_iron.append(pd.read_csv(f"Data/iron_{division}_user.csv", encoding = 'cp949'))
data_iron[0]
```

```python
data_iron[1]
```

```python
data_iron[2]
```

```python
data_iron[3]
```

모든 속성이 NaN인 행 삭제가 필요할 것으로 보임

### 결측치 처리
---
조건 : 모든 속성이 NaN인 행 
[전처리]
* 결측값이 pandas에서 제거할 수 있는 np.nan이 아닌 "NaN" 이었기 때문에 치환
* 아이언_I : 다른 티어에는 없는 컬럼 삭제

#### 아이언_I 티어 소환사 정보

```python
# 컬럼 제거
data_iron[0] = data_iron[0].drop(['miniSeries.target', 'miniSeries.wins', 'miniSeries.losses', 'miniSeries.progress'], axis = 1)
data_iron[0].info()
```

```python
data_iron[0].replace('NaN', np.nan)
```

결측치 제거

```python
data_iron[0] = data_iron[0].dropna()
data_iron[0]
```

#### 아이언_II 티어 전처리

```python
data_iron[1] = data_iron[1].replace("NaN", np.nan)
data_iron[1]
```

```python
data_iron[1] = data_iron[1].dropna()
data_iron[1]
```

#### 아이언_III 티어 소환사 정보 전처리

```python
data_iron[2] = data_iron[2].replace("NaN", np.nan)
data_iron[2]
```

```python
data_iron[2] = data_iron[2].dropna()
data_iron[2]
```

#### 아이언_IV 티어 전처리

```python
data_iron[3] = data_iron[3].replace("NaN", np.nan)
data_iron[3]
```

```python
data_iron[3] = data_iron[3].dropna()
data_iron[3]
```

### 데이터 병합

```python
df_iron = pd.concat([data_iron[0], data_iron[1], data_iron[2], data_iron[3]])
df_iron
```

## 총 경기 수 컬럼 추가

```python
df_iron['total'] = df_iron['wins'] + df_iron['losses']
df_iron
```

## 상위 5000명의 데이터 추출
---
총 플레이 한 판 수로 정렬한 뒤, 상위 5000명 데이터 추출

```python
df_iron = df_iron.sort_values(by = 'total', ascending = False)
df_iron
```

```python
df_iron = df_iron.head(6000)
df_iron.shape
```

```python
df_iron.to_csv("preprocessed data/iron.csv", encoding = 'cp949')
```
