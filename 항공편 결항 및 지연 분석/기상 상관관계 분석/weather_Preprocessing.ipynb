### 풍향 데이터 전처리

```python
import pandas as pd

wind_direction_17 = pd.read_csv("C:/Data/weather/RKPC/0/2017.csv")
wind_direction_18 = pd.read_csv("C:/Data/weather/RKPC/0/2018.csv")
wind_direction_19 = pd.read_csv("C:/Data/weather/RKPC/0/2019.csv")
wind_direction_20 = pd.read_csv("C:/Data/weather/RKPC/0/2020.csv")
wind_direction_21 = pd.read_csv("C:/Data/weather/RKPC/0/2021.csv")
print("[원본 데이터 - 2017]")
print(wind_direction_17.shape)
print("[원본 데이터 - 2018]")
print(wind_direction_18.shape)
print("[원본 데이터 - 2019]")
print(wind_direction_19.shape)
print("[원본 데이터 - 2020]")
print(wind_direction_20.shape)
print("[원본 데이터 - 2021]")
print(wind_direction_21.shape)
```

### 최저운고 데이터 전처리

```python
import pandas as pd
import math
m_17 = pd.read_csv("C:/Data/weather/RKPC/4/2017.csv")
m_18 = pd.read_csv("C:/Data/weather/RKPC/4/2018.csv")
m_19 = pd.read_csv("C:/Data/weather/RKPC/4/2019.csv")
m_20 = pd.read_csv("C:/Data/weather/RKPC/4/2020.csv")
m_21 = pd.read_csv("C:/Data/weather/RKPC/4/2021.csv")
print("[원본 데이터 - 2017]")
print(m_17.shape)
print("[원본 데이터 - 2018]")
print(m_18.shape)
print("[원본 데이터 - 2019]")
print(m_19.shape)
print("[원본 데이터 - 2020]")
print(m_20.shape)
print("[원본 데이터 - 2021]")
print(m_21.shape)
```

```python
m_17 = m_17.fillna(method = 'bfill')
m_18 = m_18.fillna(method = 'bfill')
m_19 = m_19.fillna(method = 'bfill')
m_20 = m_20.fillna(method = 'bfill')
m_21 = m_21.fillna(method = 'bfill')
m_17
```
