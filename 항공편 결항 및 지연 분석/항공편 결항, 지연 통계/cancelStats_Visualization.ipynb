## 결항 요인 분석
---
전체 공항 기준, 결항 요인의 비율 분석

```python
import pandas as pd

# 데이터 불러오기
airports = pd.read_csv("C:/Data/airport/결항통계.csv")
airports
```

```python
data = [airports['기상'].sum(), airports['A/C접속'].sum(), airports['A/C정비'].sum(), 
        airports['여객처리'].sum(), airports['복합원인'].sum(), airports['기타'].sum()]
label = ['기상', 'A/C접속', 'A/C정비', '여객처리', '복합원인', '기타']

plt.figure(figsize = (8, 8))
plt.pie(data, labels = label, autopct = '%.2f',  colors = color)
plt.legend(loc = 'upper right', fontsize = 8)
plt.show()
```

## 결항요인 분석 결과
---
* 용어
|용어|설명|
|:-:|:-|
|기상|기상에 의한 지연, 결항|
|A/C 접속|전편 항공기의 지연 및 결항이 다음 연결편에 영향이 되어 항공기가 지연, 결항되는 것|
|A/C 정비|항공기 정비로 인한 지연 및 결항|
|여객처리|승무원 휴식기간 보장 및 여객처리로 인한 지연 및 결항|
|복합원인|2가지 이상의 사유들로 인한 지연 및 결항|
|기타|지상조업, 활주로, 지상대기, 항공보안 등의 원인|
*출처 : [항공통계 작성 매뉴얼](https://www.airportal.go.kr)

* 가설 : 기상요인은 항공편 결항에 가장 많은 영향을 끼칠 것이다. 

항공편 결항요인 분석 결과, 1위는 기상요인이 72.87%로 1위로 나타났고, 2위는 기타로 나타났다. 항공편 결항에 기상요인이 가장 큰 비율을 차지할 것이라는 가설은 채택되었다. 3위는 A/C 접속으로 나타났다. 1위, 2위로 나타난 항목들과 달리 3위로 나타난 A/C 접속 항목은 인적 요인이다. 결항의 경우는 지연이 되는 경우보다 상대적으로 이용객의 불편을 더욱 많이 초래하기 때문에 해당 항목의 비율을 줄일 방안이 필요할 것으로 보인다.
