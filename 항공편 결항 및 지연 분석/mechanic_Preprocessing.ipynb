## 항공 종사자 현황 데이터 : 데이터 전처리
--- 
### 데이터 불러오기

```python
import pandas as pd

airline = pd.read_csv("C:/Data/airline/항공종사자현황.csv")
airline.head(10)
```

### 필요한 데이터만 사용

```python
airline.drop(['증감', '증감율','구성비', '누계', '누계구성비', 'LEVEL2'], axis = 1, inplace = True)
airline.head()
```

### 결측값 처리 
* 결측값이 4개 이상 있는 행 삭제

```python
airline = airline.dropna(thresh = 4)
airline
```

### 항공정비사 데이터 필터링

```python
airline_mask = (airline['LEVEL1'] == '항공정비사')&(airline['구분'] == '소계')
airline = airline[airline_mask]
airline.head()
```

### CSV 파일 저장

```python
airline.to_csv("C:/Data/airline/mechanic.csv", encoding = 'utf-8-sig')
```
